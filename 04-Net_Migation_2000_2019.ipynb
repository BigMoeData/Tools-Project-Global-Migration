```python
# Global Variables

data_file_path = 'Data'
image_file_path = 'Images'
```

```python
top_count = 15
bottom_count = 15

the_field = 'migration_count'

start_year = 2000
stop_year = 2004

the_df = pd.read_csv ( os.path.join ( data_file_path, f"Migration_{start_year}_{stop_year}.csv" ) )
the_df.sort_values ( by = the_field, ascending=False, inplace=True )

top_countries = the_df.head(top_count)
bottom_countries = the_df.tail(bottom_count)

combined_data = pd.concat([top_countries, bottom_countries])

fig, axs = plt.subplots(2, 2, figsize=(20, 20))

top_countries = the_df.head(top_count)
bottom_countries = the_df.tail(bottom_count)

combined_data = pd.concat([top_countries, bottom_countries])

bars = axs[0, 0].barh(combined_data['country'], combined_data['migration_count'],
                      color=[('red' if val < 0 else 'blue') for val in combined_data['migration_count']])
axs[0, 0].set_xticklabels(axs[0, 0].get_xticklabels(), fontsize=15)  # Adjust the fontsize
axs[0, 0].set_yticklabels(axs[0, 0].get_yticklabels(), fontsize=15)  # Adjust the fontsize
axs[0, 0].set_xlabel('Average Net Migration', fontsize=15)
axs[0, 0].set_title(f'Average Net Migration ( {start_year} - {stop_year} )', fontsize=20)
axs[0, 0].legend(handles=[bars[0], bars[-1]], labels=['Migration Inflow', 'Migration Outflow'])
axs[0, 0].invert_yaxis()  # Invert y-axis to have the highest value at the top

# Create bar chart for ave_migration_2005_to_2009

start_year = 2005
stop_year = 2009

the_df = pd.read_csv ( os.path.join ( data_file_path, f"Migration_{start_year}_{stop_year}.csv" ) )
the_df.sort_values ( by = the_field, ascending=False, inplace=True )

top_countries = the_df.head(top_count)
bottom_countries = the_df.tail(bottom_count)

combined_data = pd.concat([top_countries, bottom_countries])

bars = axs[0, 1].barh(combined_data['country'], combined_data['migration_count'],
                      color=[('red' if val < 0 else 'blue') for val in combined_data['migration_count']])
axs[0, 1].set_xticklabels(axs[0, 0].get_xticklabels(), fontsize=15)  # Adjust the fontsize
axs[0, 1].set_yticklabels(axs[0, 0].get_yticklabels(), fontsize=15)  # Adjust the fontsize
axs[0, 1].set_xlabel('Average Net Migration', fontsize=15)
axs[0, 1].set_title(f'Average Net Migration ( {start_year} - {stop_year} )', fontsize=20)
axs[0, 0].legend(handles=[bars[0], bars[-1]], labels=['Migration Inflow', 'Migration Outflow'])
axs[0, 1].invert_yaxis()  # Invert y-axis to have the highest value at the top

# Create bar chart for ave_migration_2010_to_2014

start_year = 2010
stop_year = 2014

the_df = pd.read_csv ( os.path.join ( data_file_path, f"Migration_{start_year}_{stop_year}.csv" ) )
the_df.sort_values ( by = the_field, ascending=False, inplace=True )

top_countries = the_df.head(top_count)
bottom_countries = the_df.tail(bottom_count)

combined_data = pd.concat([top_countries, bottom_countries])

bars = axs[1, 0].barh(combined_data['country'], combined_data['migration_count'],
                      color=[('red' if val < 0 else 'blue') for val in combined_data['migration_count']])
axs[1, 0].set_xticklabels(axs[0, 0].get_xticklabels(), fontsize=15)  # Adjust the fontsize
axs[1, 0].set_yticklabels(axs[0, 0].get_yticklabels(), fontsize=15)  # Adjust the fontsize
axs[1, 0].set_xlabel('Average Net Migration', fontsize=15)
axs[1, 0].set_title(f'Average Net Migration ( {start_year} - {stop_year} )', fontsize=20)
axs[0, 0].legend(handles=[bars[0], bars[-1]], labels=['Migration Inflow', 'Migration Outflow'])
axs[1, 0].invert_yaxis()  # Invert y-axis to have the highest value at the top

# Create bar chart for ave_migration_2015_to_2019

start_year = 2015
stop_year = 2019

the_df = pd.read_csv ( os.path.join ( data_file_path, f"Migration_{start_year}_{stop_year}.csv" ) )
the_df.sort_values ( by = the_field, ascending=False, inplace=True )

top_countries = the_df.head(top_count)
bottom_countries = the_df.tail(bottom_count)

combined_data = pd.concat([top_countries, bottom_countries])

bars = axs[1, 1].barh(combined_data['country'], combined_data['migration_count'],
                      color=[('red' if val < 0 else 'blue') for val in combined_data['migration_count']])
axs[1, 1].set_xticklabels(axs[0, 0].get_xticklabels(), fontsize=15)  # Adjust the fontsize
axs[1, 1].set_yticklabels(axs[0, 0].get_yticklabels(), fontsize=15)  # Adjust the fontsize
axs[1, 1].set_xlabel('Average Net Migration', fontsize=15)
axs[1, 1].set_title(f'Average Net Migration ( {start_year} - {stop_year} )', fontsize=20)
axs[0, 0].legend(handles=[bars[0], bars[-1]], labels=['Migration Inflow', 'Migration Outflow'])
axs[1, 1].invert_yaxis()  # Invert y-axis to have the highest value at the top

plt.tight_layout()

plt.savefig( os.path.join ( image_file_path, "Net_Migration_Data_2000_2019.jpg" ) )  # For JPG format
plt.savefig( os.path.join ( image_file_path, "Net_Migration_Data_2000_2019.png" ))  # For PNG format

#plt.show()
```
